```python
import requests
from bs4 import BeautifulSoup 
```

```python
#url is link to wikipedia page link
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
#Scrapping the link
res=requests.get(url)
#getting the data from responser
soup = BeautifulSoup(res.content,'lxml')
#print(soup.prettify())
```

```python
#we know the table we wanted to extract is 1st table of data
table=soup.find_all('table')[0]
#print(table.prettify)
#Converting table into dataframe
df_Canada_raw = pd.read_html(str(table),header=0)[0]
df_Canada_raw.head()
```

```python
df_Canada_raw.drop(df_Canada_raw[df_Canada_raw['Borough']=='Not assigned'].index,axis=0,inplace=True)
df_Canada_raw.reset_index()
```

```python
#df_Canada_raw.drop(columns=['index'],inplace=True)
print(df_Canada_raw.shape)
df_Canada_raw.head()
```

```python
df_Canada_raw=df_Canada_raw.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
print(df_Canada_raw.shape)
```

```python
df_Canada_raw.loc[:,'Neighbourhood'].replace('Not assigned', df_Canada_raw.loc[:,'Borough'], inplace=True)
```
